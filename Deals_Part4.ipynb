## Select deal parts with dollar value

```python
%%sql
SELECT DealName, PartNumber, DollarValue
From deals join DealParts on  (Deals.DealID = DealParts.DealID)
```

## Create a view that matches companies to deals
- Contain DealName, Rolecode, CompanyName

```python
Company_deals = %sql SELECT DealName,RoleCode,CompanyName  FROM Companies JOIN Players ON (Companies.CompanyID = Players.CompanyID) JOIN Deals ON (Players.DealID = Deals.DealID) ORDER BY DealName

print(Company_deals)
```
